## Popularity Based Recommneder System

```python
ratings
```

```python
books
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>250].sort_values('avg_rating',ascending=False).head(50)
```

```python
books
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','num_ratings','avg_rating','Book-Author','Year-Of-Publication','Publisher','Image-URL-M']]
```

```python
popular_df
```

## Collaborative Filtering Approach

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200
educated_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >50
famous_books_index = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books_index)]
```

```python
final_ratings
```

```python
final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating').fillna(0)
pt
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
pt.index[187]
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    distances = similarity_scores[index]
    similar_items = sorted(list(enumerate(distances)),key = lambda x:x[1],reverse = True)[1:6]
    data = []
    for i in similar_items:
        item=[]
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```
